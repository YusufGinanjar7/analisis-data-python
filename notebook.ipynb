```python
#cek outlier
df = pd.read_csv("data/hour.csv")
def check_outlier(df):
    print('Outlier / jumlah data\n')
    for i in df.select_dtypes(include=['int64', 'float64']).columns:
        Q1 = df[i].quantile(0.25)
        Q3 = df[i].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5*IQR
        upper = Q3 + 1.5*IQR
        print(f'Outlier {i} : {round(df[(df[i] < lower) | (df[i] > upper)].shape[0] / df.shape[0] * 100,2)}%')
check_outlier(df)
```

```python
df = pd.read_csv("data/hour.csv")

def replaceIQR(df):
    for col in df.select_dtypes(include=['int64', 'float64']).columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        
        maximum = Q3 + (1.5 * IQR)
        minimum = Q1 - (1.5 * IQR)
        
        # Mengganti nilai outlier dengan batas maksimum atau minimum
        df.loc[df[col] > maximum, col] = maximum
        df.loc[df[col] < minimum, col] = minimum


        print(f'Outlier pada kolom {col} diganti dengan batas IQR.')

replaceIQR(df)

df.to_csv("data/hour.csv", index=False)
```

```python
# Membuat fungsi untuk menentukan cluster waktu
def get_time_cluster(hr):
    if 5 <= hr <= 7:
        return 'Sunrise'
    elif 8 <= hr <= 15:
        return 'Daytime'
    elif 16 <= hr <= 19:
        return 'Sunset'
    else:
        return 'Night'

# Menambahkan kolom baru untuk cluster waktu
hour_df['time_cluster'] = hour_df['hr'].apply(get_time_cluster)

# Menampilkan beberapa baris pertama untuk memeriksa hasil
hour_df[['dteday', 'hr', 'time_cluster']].head()
```

```python
# Menghitung rata-rata peminjaman berdasarkan time_cluster
time_cluster_analysis = hour_df.groupby('time_cluster')['cnt'].mean().reset_index()

time_cluster_analysis
```

```python
# Menghitung korelasi jam dengan total peminjaman
korelasi = hour_df[['hr', 'cnt']].corr()
korelasi
```

```python
# Mengelompokkan data berdasarkan jam (hr) dan menghitung total sepeda yang dipinjam (cnt)
hourly_rentals = hour_df.groupby('hr')['cnt'].sum().reset_index()

# Mencari peminjaman paling banyak dan paling sedikit
max_rental = hour_df.loc[hour_df['cnt'].idxmax()]
min_rental = hour_df.loc[hour_df['cnt'].idxmin()]

print("\nPeminjaman Sepeda Paling Banyak:")
print(f"Waktu: {max_rental['datetime']}, Jumlah: {max_rental['cnt']}")

print("\nPeminjaman Sepeda Paling Sedikit:")
print(f"Waktu: {min_rental['datetime']}, Jumlah: {min_rental['cnt']}")

# Menampilkan data peminjaman per jam secara descending
hourly_rentals_sorted = hourly_rentals.sort_values(by='cnt', ascending=False)
print("\nData Peminjaman Sepeda per Jam (Descending):")
hourly_rentals_sorted
```

## Visualization & Explanatory Analysis

### Pertanyaan 1 : Bagaimana perbedaan antara peminjam yang sudah teregistrasi dan yang belum teregistrasi dalam meminjam sepeda dihari kerja?

```python
working_days = day_df[day_df['workingday'] == 1]

# Menghitung total peminjaman berdasarkan jenis pengguna
total_registered = working_days['registered'].sum()
total_casual = working_days['casual'].sum()

# Data untuk pie chart
labels = ['Registered', 'Casual']
sizes = [total_registered, total_casual]

# Membuat pie chart berbentuk donat dengan persentase di sebelah nama kolom
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=[f'{label} ({size / sum(sizes) * 100:.2f}%)' for label, size in zip(labels, sizes)],
        startangle=140, colors=['lightblue', 'orange'], wedgeprops={'width': 0.4})
plt.title('Perbandingan Peminjam Sepeda (Hari Kerja)')
plt.show()
```

```python
working_days = day_df[day_df['workingday'] == 1]

# Membuat line chart
plt.figure(figsize=(14, 7))
plt.plot(working_days['dteday'], working_days['registered'], label='Registered', color='blue')
plt.plot(working_days['dteday'], working_days['casual'], label='Casual', color='orange')

# Menambahkan judul dan label
plt.title('Peminjam Sepeda (Hari Kerja)')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Peminjam')
plt.xticks(rotation=45)  # Memutar label tanggal agar lebih mudah dibaca
plt.legend()

# Menampilkan plot
plt.tight_layout()
plt.show()
```

### Pertanyaan 2 : Bagaimana korelasi antara waktu matahari terbit dan terbenam dengan pola peminjaman?

```python
# Data yang telah dikelompokkan
time_cluster_analysis = hour_df.groupby('time_cluster')['cnt'].mean().reset_index()

# Membuat bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='time_cluster', y='cnt', hue='time_cluster', data=time_cluster_analysis, palette='viridis', legend=False)

# Menambahkan judul dan label
plt.title('Rata-rata Peminjaman Sepeda Berdasarkan Waktu Matahari Terbit/Terbenam')
plt.xlabel('Cluster Waktu')
plt.ylabel('Rata-rata Peminjaman (cnt)')

# Menampilkan plot
plt.show()
```

```python
time_cluster_analysis = hour_df.groupby('time_cluster')['cnt'].mean().reset_index()

# Mengurutkan data berdasarkan urutan waktu yang logis
time_cluster_analysis['time_cluster'] = pd.Categorical(time_cluster_analysis['time_cluster'],
categories=['Night', 'Sunrise', 'Daytime', 'Sunset'], ordered=True)
time_cluster_analysis = time_cluster_analysis.sort_values('time_cluster')

# Membuat line chart
plt.figure(figsize=(10, 6))
sns.lineplot(x='time_cluster', y='cnt', data=time_cluster_analysis, marker='o', color='b')

# Menambahkan judul dan label
plt.title('Perbandingan Peminjaman Sepeda Berdasarkan Waktu Matahari Terbit/Terbenam', fontsize=16)
plt.xlabel('Waktu', fontsize=14)
plt.ylabel('Rata-rata Peminjaman (cnt)', fontsize=14)

# Menampilkan grid
plt.grid(True)

# Menampilkan plot
plt.show()
```

**Insight:**
- visualisasi data pada pertanyaan pertama yaitu, bahwa peminjam yang sudah registrasi lebih banyak dibanding kan yang belum, yaitu casual. dengan nilai 83% bagi yang sudah registrasi, sedangkan yang belum, sebanyak 13%.
- visualisasi line chart untuk pertanyaan pertama menjelaskan bahwa perkembangan peminjaman sepeda untuk yang sudah registrasi dan yang belum untuk perkembangannya beda pesat, akan tetapi curva kenaikan dan turun grafik tidak terlampau jauh.
- visualisasi data pada pertanyaan kedua, untuk peminjaman sepeda paling tinggi terjadi saat sore hari. perbedaan saat malam, dan juga pagi hari dengan sore hari yaitu lebih dari 70%, sedangkan dengan siang hari lebih dari 45%.
